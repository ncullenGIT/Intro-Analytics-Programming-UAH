```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("expenditure.csv")
data
```

```python
X = data["income"].values
Y = data["expenditure"].values
n = len(X)
```

```python
plt.figure()
plt.scatter(X, Y)
plt.xlabel("Income (X)")
plt.ylabel("Expenditure (Y)")
plt.title("Scatter Plot of Expenditure vs Income")
plt.show()
```

```python
import statsmodels.api as sm
OLS = sm.OLS(data.expenditure, sm.add_constant(data.income))
lm = OLS.fit()
print(lm.summary())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('electricity.csv')
```

```python
plt.scatter(data['X'], data['Y'])
plt.xlabel('Monthly Electricity Consumption (X)')
plt.ylabel('Hourly Peak Consumption (Y)')
plt.title('Scatter Plot of Y vs X')
plt.show()
```

```python
import statsmodels.api as sm
OLS = sm.OLS(data.Y, sm.add_constant(data.X))
lm = OLS.fit()
print(lm.summary())
```
